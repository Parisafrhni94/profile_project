### 1. Find the relative proportions of movie genres represented.

```python
# import dependencies 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# create your data variabe using .read_csv 
df = pd.read_csv("IMDB-Movie-Data.csv")
# get reaquainted with the data by viewing the first 5 rows
df.head(5)
```

```python
# list the column names
df.columns
```

```python

genre_column = df['Genre']
print(genre_column)
# preview the contents of the 'Genre' column only
```

```python
# using the .value_counts() method, explore the number of each genre type in the series
# assign the result to a variable called genre_counts
genre_counts = df.Genre.value_counts()
genre_counts
```

```python
# Find the proportions of each genre by dividing your variable assigned above by the sum of 
# all of the unique genre values
# assign this to the variable genre_proportions
genre_proportion = genre_counts/genre_counts.sum()
genre_proportion
```

```python
# the length of results a above is too long to know if there are any null values present
# we can run .isna() on the series and collect the sum to check if there are missing values
# uncomment the line below to see the results
(df['Genre'].isna().sum())
```

### 4. Find the % of movies with a higher than avergage Rating

```python
# create a variable called mean_rating.
# assign mean_rating the result of running ``.mean()`` on the rating column.
mean_rating = df['Rating'].mean()

# Print the mean rating
print(mean_rating)
```

```python
# Find the dataset where a movie's rating is higher than mean_rating. 
# Divide that by the length of the total dataset.
# Calculate the mean of the 'rating' column
mean_rating = df['Rating'].mean()

# Filter the dataset for movies with rating higher than mean_rating
higher_rated_movies = df[df['Rating'] > mean_rating]

# Calculate the ratio of higher rated movies to the total dataset length
ratio = len(higher_rated_movies) / len(df)

# Print the ratio
print(ratio)
```

### 6. Make a chart to show the number of films in each rating group each year

```python
# Create a new column called rating_group
# Use to pd.cut() to bin(group) the rating_group column values by integer intervals 1-10
df['rating_group'] = pd.cut(df['Rating'], bins=range(1, 11))
```

```python
# Make a chart to show the number of films in each rating group per year
grouped_data = df.groupby(['rating_group', 'Year']).size().unstack()

# Plot the grouped data
grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6))

# Set the labels and title
plt.xlabel('Rating Group')
plt.ylabel('Number of Films')
plt.title('Number of Films in Each Rating Group per Year')

# Display the chart
plt.show()
```
